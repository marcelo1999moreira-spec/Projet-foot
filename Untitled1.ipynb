```python
# =========================================================
# 0. IMPORTS & CONFIG
# =========================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, cross_val_score, KFold, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

import joblib

sns.set(style="whitegrid")
%matplotlib inline

# For nicer display
pd.set_option("display.max_columns", 50)
pd.set_option("display.float_format", lambda x: f"{x:,.2f}")


# =========================================================
# 1. LOAD DATA (ONE TIME)
# =========================================================

df = pd.read_csv("players_21.csv")

print("Shape initiale du dataset :", df.shape)
print("\nAperçu des colonnes :")
print(df.columns.tolist())

print("\nAperçu des premières lignes :")
display(df.head())


# =========================================================
# 2. CLEANING & FEATURE ENGINEERING DE BASE
# =========================================================

# --- 2.1 Garder seulement les joueurs avec une valeur connue ---

df = df.dropna(subset=["value_eur"])
print("\nShape après suppression des valeurs manquantes sur value_eur :", df.shape)

# --- 2.2 Créer main_position (première position dans player_positions) ---

def extract_main_position(pos):
    if isinstance(pos, str):
        return pos.split(",")[0].strip()
    return "Unknown"

df["main_position"] = df["player_positions"].apply(extract_main_position)

# --- 2.3 Growth potential (potential - overall) ---

df["growth_potential"] = df["potential"] - df["overall"]

# --- 2.4 Valeur en millions d'euros ---

df["value_million"] = df["value_eur"] / 1_000_000

# --- 2.5 BMI approximatif (si taille & poids dispo) ---

if "height_cm" in df.columns and "weight_kg" in df.columns:
    df["bmi"] = df["weight_kg"] / (df["height_cm"] / 100) ** 2
else:
    df["bmi"] = np.nan

# --- 2.6 Groupes d'âge (bins) ---

df["age_group"] = pd.cut(
    df["age"],
    bins=[15, 20, 25, 30, 35, 45],
    labels=["<20", "20-25", "25-30", "30-35", "35+"]
)

# --- 2.7 Indice "offensif" et "défensif" (si colonnes existent) ---

offensive_stats = [c for c in ["pace", "shooting", "passing", "dribbling"] if c in df.columns]
defensive_stats = [c for c in ["defending", "physic"] if c in df.columns]

if offensive_stats:
    df["offensive_index"] = df[offensive_stats].mean(axis=1)
else:
    df["offensive_index"] = np.nan

if defensive_stats:
    df["defensive_index"] = df[defensive_stats].mean(axis=1)
else:
    df["defensive_index"] = np.nan

print("\nColonnes après feature engineering de base :")
print(df.columns.tolist())

display(df[["short_name", "age", "overall", "potential",
            "growth_potential", "value_million", "main_position"]].head())


# =========================================================
# 3. ANALYSE EXPLORATOIRE (EDA)
# =========================================================

# --- 3.1 Distributions simples ---

numeric_for_eda = ["age", "overall", "potential", "growth_potential",
                   "value_million", "offensive_index", "defensive_index"]

numeric_for_eda = [c for c in numeric_for_eda if c in df.columns]

for col in numeric_for_eda:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col].dropna(), bins=30, kde=True)
    plt.title(f"Distribution de {col}")
    plt.xlabel(col)
    plt.ylabel("Fréquence")
    plt.show()

# --- 3.2 Relation entre rating et valeur ---

plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="overall", y="value_million", alpha=0.3)
plt.title("Valeur (M€) vs Overall")
plt.xlabel("Overall")
plt.ylabel("Valeur (M€)")
plt.show()

plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x="potential", y="value_million", alpha=0.3, color="green")
plt.title("Valeur (M€) vs Potential")
plt.xlabel("Potential")
plt.ylabel("Valeur (M€)")
plt.show()

# --- 3.3 Valeur par position principale ---

top_positions = df["main_position"].value_counts().head(12).index
df_top_pos = df[df["main_position"].isin(top_positions)]

plt.figure(figsize=(12, 6))
sns.boxplot(data=df_top_pos, x="main_position", y="value_million")
plt.xticks(rotation=45)
plt.title("Valeur (M€) par position principale (Top 12)")
plt.xlabel("Position principale")
plt.ylabel("Valeur (M€)")
plt.show()

# --- 3.5 Heatmap de corrélation ---

corr_cols = [c for c in ["age", "overall", "potential",
                         "growth_potential", "offensive_index",
                         "defensive_index", "value_million"]
             if c in df.columns]

plt.figure(figsize=(8, 6))
sns.heatmap(df[corr_cols].corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matrice de corrélation")
plt.show()


# =========================================================
# 4. PRÉPARATION POUR LE MODELING (FEATURES CONSISTANTES)
# =========================================================

# On garde une copie "modélisation" avec les colonnes importantes
base_cols_to_keep = ["short_name", "main_position", "value_million"]

numerical_cols = [
    "age", "height_cm", "weight_kg",
    "overall", "potential", "growth_potential",
    "pace", "shooting", "passing",
    "dribbling", "defending", "physic",
    "bmi", "offensive_index", "defensive_index"
]

numerical_cols = [c for c in numerical_cols if c in df.columns]
categorical_cols = ["main_position"]

df_model = df[base_cols_to_keep + numerical_cols].dropna()

print("\nShape du df_model :", df_model.shape)
print("Colonnes numériques utilisées :", numerical_cols)
print("Colonnes catégorielles utilisées :", categorical_cols)

# --- Définition de X, y, et noms (UNE SEULE FOIS) ---

X = df_model[numerical_cols + categorical_cols]
y = df_model["value_million"]
player_names = df_model["short_name"]

print("\nShape X :", X.shape)
print("Shape y :", y.shape)

# --- Train / Test split (UNE SEULE FOIS) ---

X_train, X_test, y_train, y_test, names_train, names_test = train_test_split(
    X, y, player_names,
    test_size=0.2,
    random_state=42
)

print("\nTrain shapes :", X_train.shape, y_train.shape)
print("Test shapes  :", X_test.shape, y_test.shape)

# --- Préprocesseur commun (StandardScaler + OneHot) ---

preprocessor = ColumnTransformer(
    transformers=[
        ("num", StandardScaler(), numerical_cols),
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_cols)
    ]
)

# On ne fait PAS le scaling à la main, il sera fait dans les pipelines


# =========================================================
# 5. DÉFINITION DES MODÈLES
# =========================================================

models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(alpha=1.0),
    "Lasso Regression": Lasso(alpha=0.0005, max_iter=10000),
    "ElasticNet": ElasticNet(alpha=0.0005, l1_ratio=0.5, max_iter=10000),
    "Decision Tree": DecisionTreeRegressor(max_depth=8, random_state=42),
    "Random Forest": RandomForestRegressor(
        n_estimators=200,
        max_depth=20,
        random_state=42,
        n_jobs=-1
    ),
    "Gradient Boosting": GradientBoostingRegressor(
        n_estimators=200,
        learning_rate=0.05,
        max_depth=3,
        random_state=42
    )
}


# =========================================================
# 6. FONCTIONS D'ÉVALUATION
# =========================================================

def compute_metrics(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)
    return mae, rmse, r2


def plot_real_vs_pred(y_true, y_pred, title="Real vs Predicted"):
    plt.figure(figsize=(7, 5))
    plt.scatter(y_true, y_pred, alpha=0.4)
    min_val = min(y_true.min(), y_pred.min())
    max_val = max(y_true.max(), y_pred.max())
    plt.plot([min_val, max_val], [min_val, max_val], "r--", linewidth=2)
    plt.xlabel("Valeur réelle (M€)")
    plt.ylabel("Valeur prédite (M€)")
    plt.title(title)
    plt.show()


# =========================================================
# 7. ENTRAÎNER ET ÉVALUER TOUS LES MODÈLES + CROSS-VAL
# =========================================================

results = []

kf = KFold(n_splits=5, shuffle=True, random_state=42)

best_model_name = None
best_rmse = np.inf
best_pipe = None

for name, model in models.items():
    print(f"\n===== Modèle : {name} =====")

    pipe = Pipeline(steps=[
        ("preprocessor", preprocessor),
        ("model", model)
    ])

    # Cross-validation sur le train
    cv_scores = cross_val_score(
        pipe, X_train, y_train,
        cv=kf,
        scoring="neg_root_mean_squared_error",
        n_jobs=-1
    )
    cv_rmse_mean = -cv_scores.mean()
    cv_rmse_std = cv_scores.std()

    print(f"CV RMSE (moyenne) : {cv_rmse_mean:.4f} M€ (+/- {cv_rmse_std:.4f})")

    # Fit sur le train
    pipe.fit(X_train, y_train)
    y_pred_test = pipe.predict(X_test)

    mae, rmse, r2 = compute_metrics(y_test, y_pred_test)
    print(f"Test MAE  : {mae:.4f} M€")
    print(f"Test RMSE : {rmse:.4f} M€")
    print(f"Test R²   : {r2:.4f}")

    results.append({
        "model": name,
        "cv_rmse_mean": cv_rmse_mean,
        "cv_rmse_std": cv_rmse_std,
        "test_mae": mae,
        "test_rmse": rmse,
        "test_r2": r2
    })

    # Garder le meilleur (sur RMSE test)
    if rmse < best_rmse:
        best_rmse = rmse
        best_model_name = name
        best_pipe = pipe
        best_y_pred = y_pred_test

    # Graph real vs pred pour chaque modèle
    plot_real_vs_pred(y_test, y_pred_test, title=f"{name} : Réel vs Prédit")

# Résumé des résultats
results_df = pd.DataFrame(results)
print("\n=== Résultats comparatifs ===")
display(results_df.sort_values("test_rmse"))

print(f"\n>>> Meilleur modèle (RMSE) : {best_model_name} avec RMSE = {best_rmse:.4f} M€")


# =========================================================
# 8. LIGHT GRIDSEARCH SUR LE MEILLEUR TYPE DE MODÈLE (EX: RANDOM FOREST)
# =========================================================

# On fait du tuning uniquement si le meilleur est Random Forest
tuned_pipe = best_pipe
tuning_done = False

if best_model_name == "Random Forest":
    print("\nGridSearchCV sur Random Forest...")

    param_grid = {
        "model__n_estimators": [200, 400],
        "model__max_depth": [15, 20, None],
        "model__min_samples_split": [2, 5],
        "model__min_samples_leaf": [1, 2]
    }

    rf_pipe = Pipeline(steps=[
        ("preprocessor", preprocessor),
        ("model", RandomForestRegressor(random_state=42, n_jobs=-1))
    ])

    grid_search = GridSearchCV(
        rf_pipe,
        param_grid,
        cv=3,
        scoring="neg_root_mean_squared_error",
        n_jobs=-1,
        verbose=1
    )

    grid_search.fit(X_train, y_train)

    print("\nMeilleurs paramètres RandomForest :")
    print(grid_search.best_params_)

    tuned_pipe = grid_search.best_estimator_
    tuned_y_pred = tuned_pipe.predict(X_test)

    tuned_mae, tuned_rmse, tuned_r2 = compute_metrics(y_test, tuned_y_pred)

    print("\nPerformance après tuning :")
    print(f"MAE  : {tuned_mae:.4f} M€")
    print(f"RMSE : {tuned_rmse:.4f} M€")
    print(f"R²   : {tuned_r2:.4f}")

    plot_real_vs_pred(y_test, tuned_y_pred, title="Random Forest (tuned) : Réel vs Prédit")

    # On remplace le best_pipe par le tuned
    best_pipe = tuned_pipe
    best_y_pred = tuned_y_pred
    best_rmse = tuned_rmse
    best_model_name = "Random Forest (tuned)"
    tuning_done = True

print(f"\n>>> Modèle final retenu : {best_model_name} (RMSE = {best_rmse:.4f} M€)")


# =========================================================
# 9. ANALYSE DES ERREURS (BEST MODEL)
# =========================================================

errors_df = pd.DataFrame({
    "short_name": names_test.values,
    "real_value_M": y_test.values,
    "pred_value_M": best_y_pred,
})

errors_df["error_M"] = errors_df["pred_value_M"] - errors_df["real_value_M"]
errors_df["abs_error_M"] = errors_df["error_M"].abs()

print("\nAperçu des erreurs :")
display(errors_df.head())

# Top 15 joueurs sous-estimés (valeur réelle > prédite)
under_valued = errors_df.sort_values("error_M", ascending=True).head(15)
print("\nTop 15 joueurs sous-estimés (modèle prédit trop bas) :")
display(under_valued)

# Top 15 joueurs sur-estimés
over_valued = errors_df.sort_values("error_M", ascending=False).head(15)
print("\nTop 15 joueurs sur-estimés (modèle prédit trop haut) :")
display(over_valued)

# Distribution des erreurs
plt.figure(figsize=(8, 4))
sns.histplot(errors_df["error_M"], bins=40, kde=True)
plt.title("Distribution des erreurs (Valeur prédite - réelle, en M€)")
plt.xlabel("Erreur (M€)")
plt.ylabel("Fréquence")
plt.show()

# Erreur en fonction de la valeur réelle
plt.figure(figsize=(8, 5))
plt.scatter(errors_df["real_value_M"], errors_df["error_M"], alpha=0.4)
plt.axhline(0, color="red", linestyle="--")
plt.xlabel("Valeur réelle (M€)")
plt.ylabel("Erreur (M€)")
plt.title("Erreur de prédiction en fonction de la valeur réelle")
plt.show()


# =========================================================
# 10. IMPORTANCE DES FEATURES (SI MODÈLE ARBRE)
# =========================================================

# On essaie d'extraire l'importance des features pour un RandomForest ou GradientBoosting

tree_model = None

if "Random Forest" in best_model_name:
    tree_model = best_pipe.named_steps["model"]
elif best_model_name == "Gradient Boosting":
    tree_model = best_pipe.named_steps["model"]

if tree_model is not None:
    # Récupérer les noms des features après le préprocessing
    ohe = best_pipe.named_steps["preprocessor"].named_transformers_["cat"]
    num_features = numerical_cols
    cat_feature_names = list(ohe.get_feature_names_out(categorical_cols))
    all_feature_names = num_features + cat_feature_names

    importances = tree_model.feature_importances_
    feat_imp = pd.DataFrame({
        "feature": all_feature_names,
        "importance": importances
    }).sort_values("importance", ascending=False)

    print("\nImportance des features (top 20) :")
    display(feat_imp.head(20))

    plt.figure(figsize=(10, 6))
    sns.barplot(data=feat_imp.head(20), x="importance", y="feature")
    plt.title(f"Top 20 des features les plus importantes ({best_model_name})")
    plt.xlabel("Importance")
    plt.ylabel("Feature")
    plt.show()
else:
    print("\nPas de modèle de type arbre sélectionné, impossible de tracer l'importance des features.")


# =========================================================
# 11. SAUVEGARDE DU MODÈLE FINAL
# =========================================================

joblib.dump(best_pipe, "player_value_model.pkl")
print("\n✅ Modèle final sauvegardé dans 'player_value_model.pkl'")


# =========================================================
# 12. FONCTIONS INTERACTIVES : PRÉDIRE UN JOUEUR
# =========================================================

# On gardera df_model comme base de données pour les joueurs
print("\nAperçu df_model (base des joueurs pour prédiction) :")
display(df_model.head())


def get_player_row(player_name, df_source):
    """
    Retourne la ligne correspondant à un joueur par son nom,
    ou None si non trouvé.
    """
    mask = df_source["short_name"].str.lower() == player_name.lower()
    if mask.sum() == 0:
        return None
    # En cas de doublons, on prend le premier
    return df_source[mask].iloc[0]


def predict_player_value(player_name, df_source, pipeline, num_cols, cat_cols):
    """
    Prédit la valeur d'un joueur en M€ à partir de son nom.
    """
    row = get_player_row(player_name, df_source)
    if row is None:
        print(f"❌ Joueur '{player_name}' introuvable dans le dataset.")
        return

    # Construire X pour ce joueur (1 ligne)
    data = pd.DataFrame([row[num_cols + cat_cols].values], columns=num_cols + cat_cols)

    pred = pipeline.predict(data)[0]

    print(f"\n🎯 Prédiction pour {player_name} :")
    print(f"Valeur prédite : {pred:.2f} M€")

    # Si la valeur réelle existe, on l'affiche pour comparaison
    if "value_million" in row:
        real = row["value_million"]
        print(f"Valeur réelle  : {real:.2f} M€")
        print(f"Erreur         : {pred - real:.2f} M€")


def compare_two_players(name1, name2, df_source, pipeline, num_cols, cat_cols):
    """
    Compare deux joueurs en affichant leurs stats clés et la valeur prédite.
    """
    row1 = get_player_row(name1, df_source)
    row2 = get_player_row(name2, df_source)

    if row1 is None:
        print(f"❌ Joueur '{name1}' introuvable.")
        return
    if row2 is None:
        print(f"❌ Joueur '{name2}' introuvable.")
        return

    data = pd.DataFrame(
        [row1[num_cols + cat_cols].values, row2[num_cols + cat_cols].values],
        columns=num_cols + cat_cols
    )

    preds = pipeline.predict(data)

    print(f"\n⚔️ Comparaison entre {name1} et {name2} :")

    for i, (name, row, pred) in enumerate(zip([name1, name2], [row1, row2], preds)):
        print(f"\n{name} :")
        print(f" - Position : {row['main_position']}")
        print(f" - Age      : {row['age']}")
        print(f" - Overall  : {row['overall']}")
        print(f" - Potential: {row['potential']}")
        if "offensive_index" in row:
            print(f" - Offensive index: {row['offensive_index']:.2f}")
        if "defensive_index" in row:
            print(f" - Defensive index: {row['defensive_index']:.2f}")
        print(f" - Valeur prédite : {pred:.2f} M€")
        if "value_million" in row:
            print(f" - Valeur réelle  : {row['value_million']:.2f} M€")

    # Indiquer lequel a la valeur prédite la plus haute
    if preds[0] > preds[1]:
        print(f"\n👉 D'après le modèle, {name1} a une valeur plus élevée.")
    elif preds[1] > preds[0]:
        print(f"\n👉 D'après le modèle, {name2} a une valeur plus élevée.")
    else:
        print("\n👉 D'après le modèle, les deux joueurs ont la même valeur prédite.")


# =========================================================
# 13. EXEMPLES D'UTILISATION INTERACTIVE
# =========================================================

# ⚠️ Mets ici des noms qui existent bien dans players_21.csv
# Exemple : "L. Messi", "Cristiano Ronaldo", etc. (vérifie dans df_model['short_name'])

print("\n=== Exemples de prédiction pour quelques joueurs ===")

example_names = df_model["short_name"].sample(3, random_state=1).tolist()
print("Joueurs pris au hasard pour test :", example_names)

for name in example_names:
    predict_player_value(name, df_model, best_pipe, numerical_cols, categorical_cols)

print("\n=== Exemple de comparaison entre deux joueurs ===")
if len(df_model) >= 2:
    name1, name2 = df_model["short_name"].iloc[0], df_model["short_name"].iloc[1]
    compare_two_players(name1, name2, df_model, best_pipe, numerical_cols, categorical_cols)

# Tu peux aussi appeler manuellement :
# predict_player_value("L. Messi", df_model, best_pipe, numerical_cols, categorical_cols)
# compare_two_players("L. Messi", "Cristiano Ronaldo", df_model, best_pipe, numerical_cols, categorical_cols)
```
